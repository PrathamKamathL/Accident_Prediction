```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
```

```python
data_accidents = pd.read_csv('data/AccidentsBig.csv')
```

```python
data_accidents
```

```python
data_casualities = pd.read_csv('data/CasualtiesBig.csv')
```

```python
data_casualities
```

```python
data_vehicle = pd.read_csv('data/VehiclesBig.csv')
```

```python
data_vehicle
```

```python
#Here we can observe that Accident_index can be used to merge.
#Lot of feature engineering is required. Let's seeeee
df = pd.merge(data_casualities, data_accidents, how='inner', on='Accident_Index')
data = pd.merge(df, data_vehicle, on='Accident_Index', how='inner')
```

```python
data
```

```python
data.info()
```

```python
#Thank God. We have only few Null values. Still I'll confirm
print(data.isnull().sum())
```

```python
data.dropna(subset=['LSOA_of_Accident_Location', 'Time'], inplace=True)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#Thank god. No more Null values
data.columns
```

```python
#We can see that the featurs in the merged df are not important. So I'll remove them
cols_to_drop = [
    'Accident_Index',
    'LSOA_of_Accident_Location',
    'Time',
    'Date',
    'longitude',
    'latitude',
    'Vehicle_Reference_y',
    'Vehicle_Reference_x',
    'Casualty_Reference'
]
data.drop(columns=cols_to_drop, inplace=True)
```

```python
data
```

```python
data.drop(columns='Local_Authority_(Highway)', inplace=True)
```

```python
data
```

```python
x = data.drop('Accident_Severity', axis = 1)
y = data['Accident_Severity']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42, stratify=y)
```

```python
y_test
```

```python
y_train
```

```python
data.info()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
pipe = Pipeline(
    [
        ('scaler', StandardScaler()),
        ('classifier', LogisticRegression(
            max_iter=1000,
            class_weight='balanced',
        ))
    ]
)
pipe.fit(x_train, y_train)
```

```python
random_forest = RandomForestClassifier(
    n_estimators=200,
    max_depth=20,
    class_weight='balanced',
    random_state=42
)
gradient = GradientBoostingClassifier(
    n_estimators=50,
    learning_rate=0.1,
    max_depth=6,
    random_state=42
)
ada = AdaBoostClassifier(
    n_estimators=200,
    learning_rate=0.1,
    random_state=42
)
xgb = XGBClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=10,
    random_state=42
)
```

```python
random_forest.fit(x_train, y_train)
```

```python
gradient.fit(x_train, y_train)
```

```python
ada.fit(x_train, y_train)
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
y_train_conv = label.fit_transform(y_train)
y_test_conv = label.transform(y_test)
xgb.fit(x_train, y_train_conv)
```

```python
from sklearn.metrics import classification_report
models = {
    "logistic regression": pipe,
    "gradient boosting": gradient,
    "random forest": random_forest,
    "ada boosting": ada,
}
for name, model in models.items():
    y_pred = model.predict(x_test)
    print(name)
    print(classification_report(y_test, y_pred))
    print("============")
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)

x_train_res, y_train_res = smote.fit_resample(
    x_train, y_train
)
```

```python
logistic_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', LogisticRegression(
        max_iter=1000,
        class_weight='balanced'
    ))
])

logistic_pipe.fit(x_train_res, y_train_res)
```

```python
random_forest = RandomForestClassifier(
    n_estimators=300,
    max_depth=20,
    class_weight='balanced',
    random_state=42
)

random_forest.fit(x_train_res, y_train_res)
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier
gradient = HistGradientBoostingClassifier(
    max_depth=6,
    learning_rate=0.1,
    class_weight='balanced',
    random_state=42
)

gradient.fit(x_train_res, y_train_res)
```

```python
ada = AdaBoostClassifier(
    n_estimators=300,
    learning_rate=0.1,
    random_state=42
)

ada.fit(x_train_res, y_train_res)
```

```python
from sklearn.utils.class_weight import compute_sample_weight
label_encoder = LabelEncoder()
y_train_enc = label_encoder.fit_transform(y_train)
y_test_enc = label_encoder.transform(y_test)

sample_weights = compute_sample_weight(
    class_weight='balanced',
    y=y_train_enc
)

xgb = XGBClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=8,
    objective='multi:softprob',
    num_class=3,
    eval_metric='mlogloss',
    random_state=42
)

xgb.fit(
    x_train,
    y_train_enc,
    sample_weight=sample_weights
)
```

```python
models = {
    "Logistic Regression": logistic_pipe,
    "Random Forest": random_forest,
    "Gradient Boosting": gradient,
    "AdaBoost": ada
}

for name, model in models.items():
    y_pred = model.predict(x_test)
    print(f"\n{name}")
    print(classification_report(y_test, y_pred))
    print("========")

y_pred_xgb_enc = xgb.predict(x_test)
y_pred_xgb = label_encoder.inverse_transform(y_pred_xgb_enc)

print("\nXGBOOST")
print(classification_report(y_test, y_pred_xgb))
print("=======")
```
